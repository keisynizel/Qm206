```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([0,2,4,1],[3,1,7,4],color="crimson",linewidth=3,alpha=0.9,marker="o",markersize=15,markeredgecolor="black",markerfacecolor="purple",markeredgewidth=5)
plt.xlabel("$x(cm²)$")
plt.ylabel("$Intensidad(\mu)$")
plt.title("Experimento No. 1")
plt.grid(True)
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(4*x)
y2=np.cos(2*x)
plt.plot(x,y,color="mediumpurple")
plt.plot(x,y2,color="black",dashes=[8,2,2,2])
plt.title("Función Sen(4x) 0-$2\pi$")
plt.xlabel("t(seg)")
plt.ylabel("Amplitud de la onda")
plt.grid(True)
plt.show()
```

```python
x=np.random.rand(100)*10
y=np.random.rand(100)*10
r=np.random.rand(100)*250+50
plt.scatter(x,y,s=r,color="yellowgreen")
plt.show()
```

```python
x=np.random.randn(10000)
plt.hist(x,bins=50,facecolor="cyan",edgecolor="black",density=True)
plt.title("Distribución de la población")
plt.ylabel("Densidad habitantes/$m²$")
plt.text(-3,0.35, r'$\mu=0,\ \sigma=15$')
plt.show()
```
