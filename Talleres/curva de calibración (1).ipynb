```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class RegLin:
    def __init__(self):
        self.pendiente=0
        self.intercepto=0
        """Calcula la pendiente y el intercepto de la regresi√≥n lineal"""
    def regresion_lineal(self,x,y):
        n=len(x)
        detM=n*sum(x*x)-sum(x)*sum(x)
        detA=sum(y)*sum(x*x)-sum(x*y)*sum(x)
        detB=n*sum(x*y)-sum(x)*sum(y)
        a=detA/detM
        b=detB/detM
        self.pendiente=b
        self.intercepto=a
        return [a,b]
    def graficar(self,x,y):
        ycalc=x*self.pendiente+self.intercepto
        plt.plot(x,y,'ro')
        plt.plot(x,ycalc,'b--')
        plt.xlabel("t(seg)")
        plt.ylabel("distancia(m)")
        plt.title("Distancia vs tiempo")
        plt.show()
```
