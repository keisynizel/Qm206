# Fórmulas Químicas

```python
formula="Sb2O5"
n=len(formula)
elementos=[]
for i in range(n):
    print(formula[i])
    if formula[i]>="A" and formula[i]<="Z":
        elementos.append(formula[i])
    if formula[i]>="a" and formula[i]<="z":
        s=elementos.pop()
        s=s+formula[i]
        elementos.append(s)
print(elementos)
```
